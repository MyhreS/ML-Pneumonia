After augmenting images, this will view the result.

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import glob
import random
```

```python
# Get paths to data
train_normal_paths = glob.glob("../chest-xray-augmented/train/NORMAL/*")
train_pneumonia_paths = glob.glob("../chest-xray-augmented/train/PNEUMONIA/*")
print("Length train normal:", len(train_normal_paths))
print("Length train pneumonia:", len(train_normal_paths))

test_normal_paths = glob.glob("../chest-xray-augmented/test/NORMAL/*")
test_pneumonia_paths = glob.glob("../chest-xray-augmented/test/PNEUMONIA/*")
print("Length test normal:", len(test_normal_paths))
print("Length test pneumonia:", len(test_pneumonia_paths))

val_normal_paths = glob.glob("../chest-xray-augmented/val/NORMAL/*")
val_pneumonia_paths = glob.glob("../chest-xray-augmented/val/PNEUMONIA/*")
print("Length val normal:", len(val_normal_paths))
print("Length val pneumonia:", len(val_pneumonia_paths))
```

Looking at the paths, we can se that there have been created many more images by augmentation. The amount of normal and pneumonia is the same, which suggest the augmentation has worked as ecpected.

We now have around 110 000 images to train on, and 10 000 images to validate on.


```python
def read_images_from_path(images_path):
    images = []
    for image_path in images_path:
        images.append(mpimg.imread(image_path))
    return np.array(images)

# Read 9 validation images and view them
sample_paths = random.choices(val_normal_paths, k=9)
sample_images = read_images_from_path(sample_paths)

plt.figure(figsize=(10,10))
for i in range(9):
    ax = plt.subplot(3,3, i+1)
    plt.imshow(sample_images[i], cmap="gray")
    plt.axis("off")
```

From viewing the images we can see they have been augmented. The point of augmenting is to create more data so the model can generalize. The amount of augmentation may need to be changed if it is either augmened too much or too little. If that is the case the model will have problems.
